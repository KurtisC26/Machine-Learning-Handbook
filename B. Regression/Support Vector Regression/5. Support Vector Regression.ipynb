# Support Vector Regression Theory
<br>
In a normal regression model, we will try to minimze the sum of squares in a method called ordinary least squares. Within SVR, we will plot a regression line within a Epsolon Insentivie Tube which provides a margin of error describe by the epsolon above and below the line. We then want to minimize the error between the outsie points and the tube.
<br>
<br>
There are other models, this example and formula is for linear SVR
<br>







### Importing and processing data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = pd.read_csv("Position_Salaries_5.csv")
```

```python
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values

y = y.reshape(len(y),1)


# Feature scale the y variables:
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
sc_y = StandardScaler()

x = sc_x.fit_transform(x)
y = sc_y.fit_transform(y)


```

```python
# Training the SVR model:

from sklearn.svm import SVR
model = SVR(kernel = 'rbf')
model.fit(x,y)


sc_y.inverse_transform(model.predict(sc_x.transform([[6.5]])))

```

```python
#visualising SVR Results:
plt.scatter(sc_x.inverse_transform(x),sc_y.inverse_transform(y),color='red')

plt.plot(sc_x.inverse_transform(x),sc_y.inverse_transform(model.predict(x)),color='blue')
plt.title("Salaries vs SVR regression model")
plt.xlabel("Position Level")
plt.ylabel("Salary")
```
