```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset = pd.read_csv("Salary_Data_2.csv")
```

### In this case we are trying to predict the salary of employees given the number of years of experience

```python
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

### Training a simple regrression model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
model = regressor.fit(x_train,y_train)
```

```python
y_preds = model.predict(x_test)
```

### Visualising the prediction resulsts

```python
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,model.predict(x_train))
plt.title("Predicted Salary vs Actual on Trainig set")
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.show()
```

```python
plt.scatter(x_test,y_test,color='red')
plt.plot(x_train,model.predict(x_train))
plt.title("Predicted Salary vs Actual on Test set")
plt.xlabel("Years of experience")
plt.ylabel("Salary")
plt.show()
```
