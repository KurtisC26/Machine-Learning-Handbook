```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Position_Salaries_4.csv")
```

```python
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_features = PolynomialFeatures(degree=4)
x_ploy = poly_features.fit_transform(x)

lin_reg_2 = LinearRegression()
lin_reg_2.fit(x_ploy,y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue')
plt.title("Salaries vs linear regression model")
plt.xlabel("Position Level")
plt.ylabel("Salary")


```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg_2.predict(x_ploy),color='blue')
plt.title("Salaries vs Polynomial regression model")
plt.xlabel("Position Level")
plt.ylabel("Salary")

```

```python
# PRedicting a new salary result with both models

lin_reg.predict([[6.5]])
```

```python
lin_reg_2.predict(poly_features.fit_transform([[6.5]]))
```
