# Decision Tree Regression
<br>
More useful for high-dimension data sets (many varibales)
<br>
Regression trees are more complex than classification trees. The splits are determined by the algorithm through a concept called information entropy (complex math concept). Each split would be considered a terminal leaf in a more traditional style of a decision tree. The end result in a graph could look like this:







```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Position_Salaries_6.csv")
x = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.tree import DecisionTreeRegressor

model = DecisionTreeRegressor()
```

```python
model.fit(x,y)
```

```python
model.predict([[6.5]])
```

```python
X_grid = np.arange(min(x), max(x), 0.01)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(x, y, color = 'red')
plt.plot(X_grid, model.predict(X_grid), color = 'blue')
plt.title('Truth or Bluff (Decision Tree Regression)')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.show()
```
